```python
import torch
import pandas as pd
import numpy as np

import plotly.graph_objects as go
```

```python
data_path = "/home/preston/ADAM/ADAM-Logging/logs/hzd_simulation_learning_data.csv"
df = pd.read_csv(data_path)
df[["time", "stance_foot"]].to_numpy()
```

```python
import pandas as 
```

```python
# Pull out stance foot data.
stance_foot = np.array(df["stance_foot"])

# Find preimpact indices i.e., index of all leading/trailing edges in the data.
preimpact_indices = np.where(np.diff(stance_foot) != 0)[0]

df = df.iloc[preimpact_indices]
```

```python
# Wrap-safe angle difference for numpy arrays
def angle_diff(a, b):
    return (a - b + np.pi) % (2 * np.pi) - np.pi

# Finite difference the yaw angle.
yaw_rate = (
    angle_diff(df["yaw_world"].values[1:], df["yaw_world"].values[:-1])
    / df["time"].diff().values[1:]
)


# # # Plot the yaw rate.
fig = go.Figure()
fig.add_trace(go.Scatter(x=df["time"], y=yaw_rate, mode="lines"))
fig.add_trace(go.Scatter(x=df["time"], y=df["w_z_body_ctrl"], mode="lines"))
fig.update_layout(
    title="Yaw Rate",
    xaxis_title="Time (s)",
    yaxis_title="Yaw Rate (rad/s)",
)
fig.show()
```

```python
# Plot v_x_body_ctrl, v_x_body_ref of the data.
fig = go.Figure()
fig.add_trace(
    go.Scatter(x=df["time"], y=df["v_x_body_ref"], mode="lines", name="v_x_body_ref")
)
fig.add_trace(
    go.Scatter(x=df["time"], y=df["v_x_body_ctrl"], mode="lines", name="v_x_body_ctrl")
)
fig.update_layout(
    title="v_x_body_ref vs v_x_body_ctrl",
    xaxis_title="Time (s)",
    yaxis_title="Velocity (m/s)",
)
fig.show()
```

```python
# Plot v_x_body_ctrl and the actual body velocity.
fig = go.Figure()
fig.add_trace(
    go.Scatter(x=df["time"], y=df["v_x_body_ctrl"], mode="lines", name="v_x_body_ctrl")
)
fig.add_trace(go.Scatter(x=df["time"], y=df["v_x_body"], mode="lines", name="v_x_body"))
fig.update_layout(
    title="v_x_body_ctrl vs v_x_body",
    xaxis_title="Time (s)",
    yaxis_title="Velocity (m/s)",
)
fig.show()
```

```python
# Plot v_y_body_ctrl and the actual body velocity.
fig = go.Figure()
fig.add_trace(
    go.Scatter(x=df["time"], y=df["v_y_body_ctrl"], mode="lines", name="v_y_body_ctrl")
)
fig.add_trace(go.Scatter(x=df["time"], y=df["v_y_body"], mode="lines", name="v_y_body"))
fig.update_layout(
    title="v_y_body_ctrl vs v_y_body",
    xaxis_title="Time (s)",
    yaxis_title="Velocity (m/s)",
)
fig.show()
```

```python
# Plot yaw_rate and the actual body yaw rate.
fig = go.Figure()
fig.add_trace(
    go.Scatter(x=df["time"], y=df["w_z_body_ctrl"], mode="lines", name="yaw_rate_ctrl")
)
fig.add_trace(go.Scatter(x=df["time"], y=df["w_z_body"], mode="lines", name="yaw_rate"))
fig.update_layout(
    title="yaw_rate_ctrl vs yaw_rate",
    xaxis_title="Time (s)",
    yaxis_title="Velocity (m/s)",
)
fig.show()
```
